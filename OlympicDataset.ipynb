```python
plt.scatter("Height","Weight",data=data)
plt.xlabel("height")
plt.ylabel("weight")
plt.title("height vs weight")
plt.show()
```

```python
sns.set_style("darkgrid")
sns.scatterplot(x="Height",y="Weight",hue="Sex",data=data) #HUE 
plt.xlabel("Height of athlete")
plt.ylabel("Weight of athlete")
plt.title("Weight VS Height")
plt.show()
```

```python
data.columns
data["Medal"].unique()
```

```python
sns.set_style("whitegrid")
sns.scatterplot(x="Weight",y="Height",hue="Medal",data=data)
plt.show()
```

```python
sns.set_style("whitegrid")
sns.scatterplot(x="Weight",y="Height",hue="Medal",style="Sex",data=data)
plt.show()
```

```python
sns.set_style("darkgrid")
sns.lineplot(x="Height",y="Weight",hue="Sex",data=data)
plt.show()
```

```python
sns.set_style("white")
sns.displot(x="Height",hue="Sex",data=data)
plt.show()
```

```python
sns.set_style("white")
sns.displot(x="Height",hue="Sex",kind="kde",data=data)
plt.show()
```

```python
sns.set_style("whitegrid")
sns.barplot(x="Medal",y="Height",hue="Sex",data=data)
plt.show()
```

```python
sns.catplot(x="Medal",y="Height",hue="Sex",col="Season",data=data)
```
