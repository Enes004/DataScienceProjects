```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
'''
.loc → satır seçmek

.iloc → pozisyona göre satır seçmek

.max() / .min() / .mean() → basit istatistikler

.idxmax() / .idxmin() → max/min değerin satırını bulmak

df[condition] → filtreleme

'''
```

```python
data = {
"name":["Enes", "Ayşe", "Mehmet", "Fatma", "Baran", "Lara", "Okan"],
"age" :[20, 22, 19, 24, 21, 23, 20],
"salary":[12000, 15000, 11000, 18000, 12500, 17000, 14000],
"experience":[1, 2, 1, 3, 1, 2, 2]   
}

df = pd.DataFrame(data)
df
```

```python
'''
.iloc[row_position , col_position]
→ pozisyon ister (0,1,2…)

.loc[row_label , col_label]

→ isim ister (“age”, “salary”…)

'''
```

```python
fig , ax = plt.subplots(1,2, figsize=(15,5))
sns.scatterplot(x="age",y="salary",data=df,ax=ax[0])
ax[0].set_title("Age vs Salary")
ax[0].set_xlabel("Age")
ax[0].set_ylabel("Salary")
sns.barplot(x="name",y="performance",data=df,ax=ax[1])
ax[1].set_title("Performance Levels")
plt.tight_layout()
plt.show()
```

```python
fig , ax = plt.subplots(1,2, figsize=(15,5))
sns.scatterplot(x="age",y="salary",data=df,ax=ax[0])
ax[0].set_title("Age vs Salary")
ax[0].set_xlabel("Age")
ax[0].set_ylabel("Salary")

# Maksimum maaşı bul
idx = df['salary'].idxmax()   # salary sütununda max değerin satır indeksi
max_row = df.loc[idx]         # o satırı çek

# Anotasyon (ok ve yazı)
ax[0].annotate(
    f"Max: {max_row['name']} ({max_row['salary']})",
    xy=(max_row['age'], max_row['salary']),          # oku nereye koyacağız
    xytext=(max_row['age']+0.5, max_row['salary']+500), # yazının pozisyonu
    arrowprops=dict(arrowstyle="->", color='red')   # okun şekli
)

sns.barplot(x="name",y="performance",data=df,ax=ax[1])
ax[1].set_title("Performance Levels")
plt.tight_layout()
plt.show()
```
