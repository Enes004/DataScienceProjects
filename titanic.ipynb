```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = sns.load_dataset("titanic")
```

```python
data.head()
```

```python
plt.figure(figsize=(10,6))

plt.subplot(1,2,1)
sns.boxplot(x="class",y="age",data=data)
plt.title("Age by Class")
plt.xlabel("Class")
plt.ylabel("Age")

plt.subplot(1,2,2)
sns.boxplot(x="class",y="fare",data=data)
plt.title("Fare by Class")
plt.xlabel("Class")
plt.ylabel("Fare")

plt.tight_layout()#sıkısık durumda daha genis gösterebilir
plt.show()
```
